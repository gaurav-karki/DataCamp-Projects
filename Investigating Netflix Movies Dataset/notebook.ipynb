```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# Read in the Netflix CSV as a DataFrame
df = pd.read_csv("netflix_data.csv")
```

```python
typemovie = df[df['type'] == 'Movie']
```

```python
rel_1990s = typemovie[(typemovie['release_year'] >= 1990) & (typemovie['release_year'] <2000)]
display(rel_1990s.value_counts)
```

```python
plt.hist(rel_1990s['duration'], bins = 10)
plt.title("Distribution of duration")
plt.xlabel("release year")
plt.ylabel("duration")
plt.show()
```

```python
action_movies_1990s = rel_1990s[rel_1990s["genre"]=="Action"]
display(action_movies_1990s)
```

```python
short_movie_count = (action_movies_1990s["duration"]<90).sum()
print(short_movie_count)
```
